```python
from pathlib import Path
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os

%matplotlib notebook
```

```python
image_folder = Path('shuffle_images/shuffle_colorful_frames')
image_0 = cv2.imread('shuffle_images/shuffle_colorful_frames/frame_078.png')
image_size = image_0.shape[:2][::-1] # later we pass this in a func which reads column row
print(image_size)
fps = 30

images = sorted(image_folder.glob("*"))
# print(images)
```

```python
video_file = "shuffle_animation_v8.avi"
fps = 30
out = cv2.VideoWriter(video_file, cv2.VideoWriter_fourcc(*"MJPG"), fps, image_size) # frame size must be column, row

for filename in images:
    img = cv2.imread(str(filename))
    # print(img.shape)
    img = img.astype('uint8')
    out.write(img)
out.release()
cv2.destroyAllWindows()
```
